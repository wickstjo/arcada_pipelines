```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
dataset = list(pd.read_csv('../finance_historical.csv')['close'].tail(1000))
```

```python
def plot_dataset(dataset, header):
    multiplier = 0.6
    plt.figure(figsize=(16*multiplier, 9*multiplier))
    
    plt.plot(dataset, linewidth=1, marker='o', markersize=2)
    plt.xlim(left=0, right=len(dataset))

    plt.title(header)
    plt.tight_layout()
    plt.grid()

    plt.savefig(f'./{header}.png')
    plt.show()
```
